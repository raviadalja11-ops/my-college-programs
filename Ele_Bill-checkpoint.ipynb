```python
# Electricity Billing System

user = int(input("Enter Electricity Unit: "))

bill = 0

if user <= 200:
    bill = user * 6
elif user <= 250:
    bill = (200 * 6) + ((user - 200) * 7)
else:
    bill = (200 * 6) + (50 * 7) + ((user - 250) * 8)

print("Total Bill = ₹", bill)
```

```python
# Converter 

no = input("Enter Decimal No: ")

while True:
    print("\n 1.Decimal to Binary")
    print("\n 2.Decimal to Octal")
    print("\n 3.Decimal to Hexa")
    print("\n 0.Exit")

    ch = int(input("Enter Any One Choice: "))

    if ch == 1:        
        print("Binary No: ", bin(no))
    elif ch == 2:
        print("Octal No: ",  oct(no))        
    elif ch == 3:
        print("HexaDecimal No: ", hex(no))
    elif ch == 0:
        break    
```
